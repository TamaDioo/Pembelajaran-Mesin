# Tugas Praktikum
Wisconsin Breast Cancer
## Deskripsi Tugas
Pada tugas pratikum ini Anda akan menggunakan data "Wisconsin Breast Cancer". Data tersebut terdiri dari 569 data yang digunakan untuk mendiagnonis jenis kanker Malignant (M) dan Benign (B). Tugas Anda adalah,
1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.
2. Lakukan proses encoding pada kolom "diagnosis".
3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
import pandas as pd

df = pd.read_csv('wbc.csv')

print(df.head())

print("\nInfo Data:")
df.info()
```

1. Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
# Membuat salinan baru untuk dimodifikasi agar data asli tetap utuh
df_processed = df.copy()

# Membuang kolom 'id' dan 'Unnamed: 32'
df_processed = df_processed.drop(columns=['id', 'Unnamed: 32'])

print("Kolom 'id' dan kolom kosong 'Unnamed: 32' telah dihapus karena tidak dapat digunakan untuk pemodelan.")
print("Kolom yang tersisa:")
print(df_processed.columns)
```

2. Lakukan proses encoding pada kolom "diagnosis".

```python
from sklearn.preprocessing import LabelEncoder

# Membuat objek LabelEncoder
le = LabelEncoder()

# Melakukan encoding pada kolom 'diagnosis'
# 'M' akan diubah menjadi 1, dan 'B' menjadi 0
df_processed['diagnosis'] = le.fit_transform(df_processed['diagnosis'])

print("Verifikasi hasil encoding kolom pada \"diagnosis\" (M=1, B=0):")
print(df_processed.head(50))

# Menghitung jumlah diagnosis 'M' (1) dan 'B' (0)
diagnosis_counts = df_processed['diagnosis'].value_counts()

print("\nJumlah Diagnosis:")
print(f"Benign (B = 0): {diagnosis_counts.get(0, 0)}")
print(f"Malignant (M = 1): {diagnosis_counts.get(1, 0)}")
```

3. Lakukan proses standardisasi pada semua kolom yang memiliki nilai numerik.

```python
from sklearn.preprocessing import StandardScaler

# Memisahkan antara fitur (X) dan target (y)
X = df_processed.drop(columns=['diagnosis'])
y = df_processed['diagnosis']

# Membuat objek StandardScaler
std = StandardScaler()

# Melakukan standarisasi pada semua kolom fitur
X_scaled = std.fit_transform(X)

# Mengubah hasil standarisasi (yang berupa array) kembali menjadi DataFrame
# agar nama kolom tidak hilang
df_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Menggabungkan kembali target (y) dengan fitur yang sudah distandarisasi (df_scaled)
df_final = pd.concat([y, df_scaled], axis=1)

print("Verifikasi hasil standarisasi:")
print(df_final.head(10))
```
